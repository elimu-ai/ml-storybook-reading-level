```python
## MIN MAX SCALER

with open("step1_2_preprocess_data.py", "r") as file:
    data = file.readlines()

# insert min-max Scaler after unnecessary columns are dropped
for i, line in enumerate(data):
    if "storybooks_dataframe = storybooks_dataframe[['id', 'reading_level'," in line:
        data.insert(i + 1, """
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
storybooks_dataframe[['chapter_count', 'paragraph_count', 'word_count', 'avg_word_length']] = scaler.fit_transform(
    storybooks_dataframe[['chapter_count', 'paragraph_count', 'word_count', 'avg_word_length']])
print(storybooks_dataframe[['chapter_count', 'paragraph_count', 'word_count', 'avg_word_length']].head())

""")

# write to file
with open("step1_2_preprocess_data.py", "w") as file:
    file.writelines(data)
```
